```python
import numpy as np
import matplotlib.pyplot as plt

# Constants (Keplerian Elements for Earth's orbit)
a = 1.0  # Semi-major axis in AU
e = 0.0167  # Eccentricity
num_points = 360  # Number of points to plot

# Define the orbit
theta = np.linspace(0, 2 * np.pi, num_points)  # Angle (true anomaly) from 0 to 360 degrees
r = a * (1 - e**2) / (1 + e * np.cos(theta))  # Polar equation of the ellipse

# Convert polar to Cartesian coordinates
x = r * np.cos(theta)
y = r * np.sin(theta)

# Plotting
plt.figure(figsize=(6,6))
plt.plot(x, y, label='Earth Orbit', color='blue')
plt.scatter(0, 0, color='orange', s=200, marker='o', label='Sun')  # Sun at the center
plt.xlabel('X [AU]')
plt.ylabel('Y [AU]')
plt.title('Earth\'s Orbit around the Sun')
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.grid(True)
plt.show()
```
